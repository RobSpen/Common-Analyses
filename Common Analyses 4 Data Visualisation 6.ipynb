# Data Visualisation 6 - Density Plots: Kernel Density Estimators (KDE)

In this notebook I demonstrate how to create density plots using matplotlib and seaborn. Kernel Density Estimators (KDE) are effective ways to plots the distribution of variables. 

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(np.random.randn(500, 4), columns = ["Mon", "Tue", "Wed", "Thur"])

df
```

```python
df.plot(kind="kde")
```

```python
df.plot.kde()
```

```python
# The density method produces the exact same plot as using a kde. 

df.plot.density()
```

```python
df.plot(kind="density")
```

```python
df.plot.kde(alpha = 0.5)
```

```python
df.plot.kde(alpha=0.5, lw=4)
```

```python
df["Mon"].plot.kde()
```

```python
df["Tue"].plot.kde()
```

```python
df["Tue"].plot.kde(title = "kde density plot for Tuesday")
```

## KDE Plots in Seaborn

```python
cars = pd.read_csv("mtcars.csv")

cars.head()
```

```python
cars.shape
```

```python
sns.kdeplot(cars.hp)
```

```python
sns.kdeplot(cars.hp, fill = True)
```

```python
sns.kdeplot(cars.hp, bw_method=20)
```

```python
sns.kdeplot(cars.hp, bw_method=10)
```

```python
sns.kdeplot(cars.hp, bw_method=0.5)
```

```python
sns.kdeplot(cars.hp, bw_method=0.2)
```

```python
sns.kdeplot(cars.hp, cumulative=True)
```

```python
sns.kdeplot(data = cars, x='hp', y='mpg')
```

```python
sns.kdeplot(data = cars, x='hp', y='mpg', n_levels=20)
```

```python
sns.kdeplot(data = cars, x='hp', y='mpg', fill=True)
```

```python
sns.kdeplot(data = cars, x='hp', y='mpg', fill=True, cbar=True)
```

```python
sns.kdeplot(data = cars, x='hp', y='mpg', hue = 'am', fill=True)
```

```python
sns.kdeplot(data = cars, x='hp', y='mpg', hue = 'cyl', fill=True)
```
