# Data Visualisation 5 - Seaborn Distribution Plot

In this notebook I will demonstrate how to create visualisation of the distribution of variables using the seaborn displot. This is a plotting method in seaborn that is very versatile and permits data to be presented as a histogram, kernel density estimator (KDE), or empirical cumulative distribution function (ecdf). Distributions of individual variables can be crearted as well as the joint distribution of two variables. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
sns.displot(data=penguins, x="flipper_length_mm")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="kde")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="ecdf")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", kde=True)
```

```python
sns.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm", kind="kde")
```

```python
g = sns.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm", kind="kde", rug=True)
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", kind="kde")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="sex", kind="kde")
```

```python
sns.displot(data=penguins, y="flipper_length_mm", hue="sex", col="species", kind="ecdf", height=4, aspect=.7)
```
