# Data Visualisation 1 - Scatterplots

In this notebook I demonstrate how to obtain some basic scatterplots for two scale variables using matplotlib and seaborn. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
comm = pd.read_excel('comm_pd_data_2022_23.xlsx')

comm.head()
```

## Scatterplot using matplotlib

Creating a simple scatter plot for two scale variables using the standard matplotlib plt.scatter() method. 

```python
# Creating a simple scatter plot between two scale variables (agreeableness and total score). 

plt.figure(figsize = (8, 5))

plt.title("Total Points scored as a function of agreeableness")

# Using the plt.scatter method. As a mimimunm I have specified the two variables to use based on axis (x, y). 
# agreeableness is the first argument (x-axis) and total_points is the second (y-axis). 
plt.scatter(comm['agreeableness'], comm['total_points'])

plt.xlabel("Agreeableness") 
plt.ylabel("Total points")

plt.show()
```

```python
plt.figure(figsize = (8, 5))
plt.style.use('ggplot') # Here using the style.use method to change to ggplot colour scheme

plt.title("Total Points scored as a function of agreeableness")

plt.scatter(comm['agreeableness'], comm['total_points'])

plt.xlabel("Agreeableness")
plt.ylabel("Total points")

plt.show() 
```

Adjusting the alpha parameter to make visualisation easier when plotting lots of data points

```python
tips = pd.read_csv('tips.csv')

tips.head()
```

```python
tips.shape
```

```python

plt.figure(figsize = (8, 5))

plt.style.use('default') # Resetting the style to default. 

plt.title("Tips as a function of total bill")

plt.scatter(tips['total_bill'], tips['tip'], alpha = 0.5)

plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

## Scatterplot using seaborn

Creating a scatter plot using the seaborn scatterplot method.

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
```

```python
sns.set_style("darkgrid")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="size", size="size")
```
