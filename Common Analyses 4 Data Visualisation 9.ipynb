# Data Visualisation 9 - Seaborn jointplot

In this notebook I demonstrate how to visualise data using seaborn's jointplot method. The jointplot allows you to plot the relationship between variables as well as the individual distribution of variables on a single plot. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
exam = pd.read_csv('Exam Anxiety.csv')

exam.head()
```

```python
# We can see that the jointplot contains a scatter showing the relationship between the x and y axis variables
# as well as histograms showing the marginal distribution of each variable. 

sns.jointplot(x=exam.Revise, y=exam.Exam)
```

```python
# Alternate syntax to obtain the same jointplot as above. 

sns.jointplot(x='Revise', y='Exam', data= exam)
```

There are six different kinds of plots that can be created using the jointplot method. The default is scatter but kde, hist, hex, reg, and resid plots are also possible using the kind argument. 

```python
# Note that the relationship plot is now a bivariate kde plot and the marginal plots are univariate kde plots. 

sns.jointplot(x='Revise', y='Exam', data= exam, kind='kde')
```

```python
# The kind='regplot' argument gives us a combination of useful plots (histograms, kde, scatter with fitted regression line). 

sns.jointplot(x='Revise', y='Exam', data= exam, kind='reg')
```

```python
# Can save the plot as a jointgrid object and then add additional elements to it. 
# Here I have overlaid a kde plot on the scatter. 

jp = sns.jointplot(x='Revise', y='Exam', data= exam)

jp.plot_joint(sns.kdeplot, color='gray', levels=5)
```

```python
sns.jointplot(x='Revise', y='Exam', data= exam, hue='Sex')
```

```python
sns.jointplot(x='Revise', y='Exam', data= exam, hue='Sex', palette='seismic')
```

```python
sns.jointplot(x='Revise', y='Exam', data= exam, hue='Sex', palette='autumn')
```

```python
# Ratio specifies the ratio of the main plot to the marginals. The default is 5:1. 
# Here i have changed this to 2:1 this makes the marginals larger relative to the relationship plot. 
sns.jointplot(x='Revise', y='Exam', data= exam, ratio=2)
```
