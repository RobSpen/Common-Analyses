```python
# 7. Convert strings to numbers

# Creating a new data frame to use for this:

df_2 = pd.DataFrame({'Col_one': ['1.1', '2.2', '3.3'], 'Col_two': ['4.4', '5.5', '6.6'], 'Col_three': ['7.7', '8.8', '-']})

df_2

# Note all the data is stored as strings. 
```

```python
df_2.dtypes

# All objects. 
```

```python
# To do maths operations on these columns we need to convert the data types to numeric. 
# We can do this using the astype method on the first two columns:

df_2.astype({'Col_one': 'float', 'Col_two': 'float'}).dtypes
```

```python
# The above method would have produced an error if you tried to use it on the third column
# because it contains a - to represent zero or a missing value. 
# Instead we use the to_numeric function on the third column and tell it to convert
# any invalid value to an NaN value. 

pd.to_numeric(df_2.Col_three, errors = 'coerce')
```

```python
# If you know the NaN values actually represent zeros, you can fill them with zeros using
# the fillna() method 

pd.to_numeric(df_2.Col_three, errors = 'coerce').fillna(0)
```

```python
# You can apply this method to the entire data frame in one line of code using the apply method:

df_2 = df_2.apply(pd.to_numeric, errors = 'coerce').fillna(0)

df_2
```

```python
# All of the data types in this data frame have now been converted to floats:

df_2.dtypes
```

```python
# 18. Aggregate by multiple functions.

# Reading in the multilevel data file. 

pre_post = pd.read_csv('New Long Multilevel Datafile.csv')
```

```python
pre_post.head()
```

```python
# Here working out the total number of rounds for the participant with PersonID == 1:

pre_post[pre_post.PersonID == 1].Rounds.sum()
```

```python
# To work out the total rounds for each participants (this is summing pre and post rounds):

pre_post.groupby('PersonID').Rounds.sum().head()
```

```python
# Not limited to aggregating by a single function such as sum. You can use the agg method and
# pass it a list of functions such as sum and count. 

# Here it is summing the number of round sfor eahc participant and the number of data points
# All the data point counts are 2 in this case because each contributes a pre and post score. 

pre_post.groupby('PersonID').Rounds.agg(['sum', 'count']).head()
```

```python
# 19. Combine the output of an aggregation with a dataframe. 

# If we canted to create a new column listing the total number of rounds for each player. 
# This was calculated using the sum method. Because sum is an aggregation function it returns 
# a reduced version of the data frame. 
```

```python
# The output for the aggregated rounds variable is smaller than the original 
# rounds variable: 

len(pre_post.groupby('PersonID').Rounds.sum())
```

```python
len(pre_post.Rounds)
```

```python
# The solution is to use the transform method which returns the same data but transforms 
# the aggreagated variable so that it is the correct shape as the original dataframe. 

total_rounds = pre_post.groupby('PersonID').Rounds.transform('sum')
len(total_rounds)
```

```python
pre_post['total_rounds'] = total_rounds
pre_post.head(10)
```

```python
# Can then work out the % of total for pre rounds:

pre_post['percent_of_total'] = pre_post.Rounds / pre_post.total_rounds

pre_post.head(10)
```
