# Exporting and Saving Plots and Figures

In this notebook I demonstrate how to export plots and figures created using matplotlib and seaborn as separate files. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
gas = pd.read_csv('gas_prices.csv')

gas.head()
```

```python
plt.figure(figsize = (8, 5))


plt.title("Gas prices changes by year (US$)")

plt.plot(gas.Year, gas.UK, label = "United Kingdom")
plt.plot(gas.Year, gas.Canada, label = "Canada")

plt.xticks(gas.Year[::3])

plt.legend()

plt.savefig('petrol_fig.png', dpi=300) # Command to save as a png file. Setting dpi (dots per inch) sets resolution. 
# The higher the dpi the more resolution. 300 is a good value. 

plt.show()
```

### Saving figures created using seaborn

```python
titan = pd.read_csv('titanic.csv')
```

```python
titan['cat_class'] = titan['Pclass'].map({1: 'First', 2: 'Second', 3: 'Third'})
```

```python
sns.violinplot(x="cat_class", y="Age", data = titan)
plt.savefig('titanic_fig.png')
```

```python
sns.violinplot(x="cat_class", y="Age", data = titan)
plt.savefig('titan_pdf_fig.pdf', dpi=300)
```

```python
sns.violinplot(x="cat_class", y="Age", data = titan)
plt.savefig('titan_tiff_fig.tiff', dpi=300)
```
