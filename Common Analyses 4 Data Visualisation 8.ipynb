# Data Visualisation 8 - Seaborn pairplot

In this notebook I demonstrate how to create pairplots using seaborn. These are useful ways to create visualisations of the distribution of variables. They allow us to obtain multiple plots with one line of code and provide plots to visualise the distributions of individual variables on the diagonal line of the pairplot grid and relationship plots between two variables on the off-diagonals. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = pd.read_csv('tips.csv')

tips.head()
```

```python
tips.info()
```

```python
sns.set_style('darkgrid')
```

```python
# Note that pairplot will create plots of all the variables in the dataframe that are measured on a scale. 
# It will ignore those variables that are objects or categorical variables. 
sns.pairplot(tips)
```

```python
# Which can change the type of plot shown on the diagonal by using the diag_kind argument.

sns.pairplot(tips, diag_kind='kde')
```

```python
# The kind argument can be used to change the relationship plots on the off-diagonals. 

sns.pairplot(tips, kind='reg')
```

```python
# We can present different plots in the upper/ lower part of the pairgrid. 
var_plots = sns.pairplot(tips)
var_plots.map_upper(sns.kdeplot, n_levels=10, color='black')
```

```python
# We can use the seaborn hue argument to separate the data by groups or categorical variables. 
# Note that the diagonal plots have changed to a default of kde plots. 
sns.pairplot(tips, hue='smoker')
```

```python
# If we are only interested in assessing certain variables we can pass those as a list using the vars argument. 
sns.pairplot(tips, hue='smoker', vars=['total_bill', 'tip'])
```

```python
sns.pairplot(tips, hue='smoker', y_vars='tip', x_vars=['total_bill', 'size'], kind='reg', height=5)
```
